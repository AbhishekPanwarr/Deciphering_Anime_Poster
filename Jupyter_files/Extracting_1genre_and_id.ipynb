```python
import pandas as pd
import time
# Replace 'your_file.xlsx' with the path to your Excel file using a raw string literal
file_path = r"C:\Users\HN035WS\Documents\Project\Jupyter_files\animes.csv"
df = pd.read_csv(file_path)
Genre = ['Sci-Fi', 'Shounen', 'Drama', 'Romance', 'Comedy', 'Hentai','Fantasy', 'Mystery']
rows_list = df.to_dict('records')
print(len(rows_list))
```

```python
def one_genre(li,Genre):
    m = len(li)
    li.reverse()
    if m>1:
        for i in range(1,m):
            if li[i].strip() in Genre:
                return li[i].strip()
            else:
                continue
    else:
        if li[0] in Genre:
            return li[0]
        else:
            return "None"
```

```python
pairs = []
for i in range(len(rows_list)):
    genre = str(rows_list[i]['genre']).split(',')
    g = one_genre(genre,Genre)
    dic = {'anime_id' : rows_list[i]['anime_id'], 'genre': g}
    pairs.append(dic)
    
print(pairs)
```

```python
import csv

# Define the field names (keys of the dictionaries)
field_names = ['anime_id', 'genre']

# Writing the list of dictionaries to the CSV file
with open('Genre.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=field_names)
    
    # Write header
    writer.writeheader()
    
    # Write rows
    for row in pairs:
        writer.writerow(row)

print("Data written to 'dataset'")
```
