```python
import pandas as pd
import requests
import os

# CSV file name
csv_file = r"C:\Users\HN035WS\Documents\Project\excels\Shounen.csv"  # Replace with your CSV file path

# Column name containing image URLs
column_name = 'img_link'  # Replace with the column containing image URLs

# Read the CSV file using pandas
try:
    df = pd.read_csv(csv_file)
except FileNotFoundError:
    print(f"File '{csv_file}' not found.")
    exit()
anime_id = list(df['anime_id'])
image_urls = list(df[column_name])
print(image_urls)
print(f'\n\n {len(anime_id)}')
```

```python
import requests
import os


# Folder to save downloaded images
output_folder = r'C:\Users\HN035WS\Documents\Project\images\Shounen'

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

def download_images(image_urls, output_folder):
    for idx, url in enumerate(image_urls):
        try:
            response = requests.get(url)
            if response.status_code == 200:
                # Extracting the filename from the URL
                filename = os.path.join(output_folder, f"{anime_id[idx]}.jpg" if url.endswith('.jpg') else f"image_{idx+1}.png")
                
                # Save the image content to a file
                with open(filename, 'wb') as f:
                    f.write(response.content)
                print(f"Image {idx+1} downloaded successfully.")
            else:
                print(f"Failed to download image {idx+1}. Status code: {response.status_code}")
        except Exception as e:
            print(f"Error downloading image {idx+1}: {e}")

# Download images from the list of URLs
download_images(image_urls, output_folder)

```
