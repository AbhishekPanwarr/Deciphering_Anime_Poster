# Removing Invalid Image types

```python
import cv2
import imghdr
```

```python
image_dir = 'images'
image_exts = ['jpg', 'jpeg', 'png', 'bmp']
```

```python
os.listdir(image_dir)
```

```python
for image_folder in os.listdir(image_dir):
    print(f'\nOpening {image_folder} folder\n')
    for image in os.listdir(os.path.join(image_dir, image_folder)):
        image_path = os.path.join(image_dir, image_folder, image)
        try:
            img = cv2.imread(image_path)
            tip = imghdr.what(image_path)
            if tip not in image_exts: 
                print(f'Image extenstion unknown {image_path}')
                os.remove(image_path)
        except Exception as e:
            print(f'Issue with the image {image_path}')
print("\nDone opening")
```
